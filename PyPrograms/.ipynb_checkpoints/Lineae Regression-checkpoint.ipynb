```python
#Implementation of linear regression using ML 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score,mean_squared_error
dataset = pd.read_csv('headbrain.csv')
x = dataset['Head Size(cm^3)'].values
y = dataset['Brain Weight(grams)'].values
x_input = x.reshape(237,1)
y_input = y.reshape(237,1)

def linear_model(x_input,y_input):
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn import linear_model
    from sklearn.metrics import r2_score,mean_squared_error
    lm = linear_model.LinearRegression()
    #model learning or model building
    lm.fit(x_input,y_input)
    y_predict = lm.predict(x_input)
    plt.scatter(x,y,color='red')
    plt.plot(x,y_predict)
    #For getting the RMSE
    print(np.sqrt(mean_squared_error(y_input,y_predict)))
    print(r2_score(y_input,y_predict))
    #To predict the value    
    y_predict = lm.predict([[4512]])
    print(y_predict)
    

    
linear_model(x_input,y_input)


```
